K Fold

```python
X = pd.dataframe(X_scaled_train)
```

```python
from sklearn.model_selection import KFold
kf =  KFold(n_splits=3)
model = LogisticRegression()

f1_scores = [] 

for i, (train_index, test_index) in enumerate(kf.split(X)):
    model.fit(X[train_index], y_train[train_index)
    prob_kfold = model.predict(X.iloc[test_index])
    pred_kfold = np.where(prob_kfold > 0.5, 1, 0)
    f1_scores.append(f1_score(y_train[test_index],pred_kfold))
```

```python
pred_train_1 = np.where(prob_train_1 > 0.5, 1, 0)
```

```python
print(classification_report(y_train, pred_train_1))
```

```python
import numpy as np
import matplotlib.pyplot as plt


fpr, tpr, thresholds = roc_curve(y_train, pred_train_1)
roc_auc = auc(fpr, tpr)

# Mirror the x-axis by converting FPR to TNR (True Negative Rate)
tnr = 1 - fpr

# Create plot
plt.figure(figsize=(10, 8))

# Plot mirrored ROC curve (TPR vs TNR)
plt.plot(tnr, tpr, color='darkorange', lw=2, 
         label=f'Mirrored ROC curve (area = {roc_auc:.3f})')

# Plot diagonal line (represents random classifier)
plt.plot([0, 1], [1, 0], color='navy', lw=2, linestyle='--')

# Set plot properties
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('True Negative Rate (1 - False Positive Rate)')
plt.ylabel('True Positive Rate')
plt.title('Mirrored Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower left")

# Display the plot
plt.grid(alpha=0.3)
plt.show()
```
